```
import pandas 

movies = pandas.read_csv("movies.csv")
credits = pandas.read_csv("credits.csv")
ratings = pandas.read_csv("ratings.csv")
```

## Calculating a weighted rating

WR = (v % (v+m)) x R + (m % (v+m)) x C

v - number of votes for a movie

m - minimum number of votes required

R - average rating of the movie

C - average rating across all movies

```
m = movies['vote_count'].quantile(0.9)
m
```

```
C = movies['vote_average'].mean()
C
```

```
movies_filtered = movies.copy().loc[movies['vote_count'] >= m]  
movies_filtered
```

```
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v/(v+m) * R) + (m/(m+v) * C)

movies_filtered['score'] = movies_filtered.apply(weighted_rating, axis=1)
movies_filtered
```

```
movies_filtered.sort_values('score', ascending=False)[['title','score']].head(10)
```

```
movies_filtered.sort_values('score', ascending=False)[['title','score']].head(10).to_dict()
```
